## Importing required Libraries

```python
import numpy as np
import pandas as pd
```

## Loading the dataset

```python
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv"
crime = pd.read_csv(url)
```

```python
crime.head()
```

```python
crime.info()
```

## Data type of each column

```python
crime.dtypes
```

## Converting dtype of Year column into datetime 

```python
crime['Year'] = pd.to_datetime(crime.Year, format='%Y')
```

```python
crime.head()
```

## Resetting index with Year column

```python
crime.set_index('Year', inplace = True)
```

```python
crime.head()
```

## Dropping Total Column

```python
crime.columns
```

```python
crime.drop('Total', axis=1, inplace=True)
```

## Groupby Year column into decades and identify the most violent decade


```python
crime_decade = crime.groupby(pd.Grouper(freq='10Y'))['Violent', 'Property', 'Murder', 'Forcible_Rape',
       'Robbery', 'Aggravated_assault', 'Burglary', 'Larceny_Theft',
       'Vehicle_Theft'].sum()
crime_decade
```

```python
crime_decade.sum(axis=1)
```

# We get to know that 2000's are the most dangerous decade to live in the US.
