```python
import numpy as np                # Multi-Dimensional array object
import pandas as pd               # Data Manipulation
import matplotlib.pyplot as plt   # Data Visualization
import seaborn as sns             # Data Visualization   
```

```python
data = pd.read_csv('USA Cars.csv')
```

```python
data.head(5)
```

```python
data.shape
```

```python
data.columns
```

```python
data.hist(figsize=(15, 15), bins=20, layout=(4, 3), color='steelblue',
                      edgecolor='firebrick', linewidth=3);
```

```python
# Boxplot to check the presence of outliers

fig, axes = plt.subplots(2, 3, figsize = (15, 8))
sns.boxplot(data = data, y = 'Engine Cylinders', ax = axes[0,0])
sns.boxplot(data = data, y = 'Engine HP', ax = axes[0,1])
sns.boxplot(data = data, y = 'highway MPG', ax = axes[1,0])
sns.boxplot(data = data, y = 'city mpg', ax = axes[1,1])
sns.boxplot(data = data, y = 'Popularity', ax = axes[0, 2])
sns.boxplot(data = data, y = 'MSRP', ax = axes[1,2]);
```

```python
# Countplot to observe the brand population

plt.figure(figsize = (12,12))
sns.set_style("white")
sns.countplot(data = data, y = "Make")
sns.set(font_scale = 1)
plt.ylabel('Make', fontsize = 18)
plt.xlabel('Count', fontsize = 18)
plt.show()
```

```python
# Bar plot to observe the brand population with Price

plt.figure(figsize = (12,12))
sns.set_style("white")
sns.barplot(data = data, y = "Make", x = "MSRP")
plt.ylabel('Make', fontsize = 20)
plt.xlabel('MSRP', fontsize = 20)
plt.show()
```

```python
# Bar plot to observe the Engine fuel type 

plt.figure(figsize = (9,6))
sns.set_style("white")
sns.countplot(data = data, y = "Engine Fuel Type")
sns.set(font_scale = 1.2)
plt.ylabel('Engine Fuel type', fontsize = 18)
plt.xlabel('Count', fontsize = 18)
plt.show()
```

```python
# Bar plot to observe the Engine fuel type with MSRP and Popularity

fig, axes = plt.subplots(2, 1, figsize = (10, 15))
sns.set_style("white")

sns.barplot(data = data, x = "MSRP", y = "Engine Fuel Type", ax = axes[0])
plt.sca(axes[0])
plt.ylabel('Engine Fuel type', fontsize = 18)
plt.xlabel('MSRP', fontsize = 18);

sns.barplot(data = data, x = "Popularity", y = "Engine Fuel Type", ax = axes[1])
plt.sca(axes[1])
plt.ylabel('', fontsize = 18)
plt.xlabel('Popularity', fontsize = 18);
```

```python
# Bar plot to observe the Engine Cylinders type with MSRP and Popularity

fig, axes = plt.subplots(1, 2, figsize = (15, 6))
sns.set_style("white")

sns.barplot(data = data, x = "Engine Cylinders", y = "MSRP", ax = axes[0])
plt.sca(axes[0])
plt.ylabel('MSRP', fontsize = 18)
plt.xlabel('Engine Cylinders', fontsize = 18);

sns.barplot(data = data, x = "Engine Cylinders", y = "Popularity", ax = axes[1])
plt.sca(axes[1])
plt.ylabel('Popularity', fontsize = 18)
plt.xlabel('Engine Cylinders', fontsize = 18);
```

```python
# Bar plot to observe the Transmission type with MSRP and Popularity

fig, axes = plt.subplots(1, 2, figsize = (15, 6))
sns.set_style("white")

sns.barplot(data = data, x = "Transmission Type", y = "MSRP", ax = axes[0])
plt.sca(axes[0])
plt.ylabel('MSRP', fontsize = 18)
plt.xlabel('Transmission Type', fontsize = 18)
plt.xticks(rotation = 15);

sns.barplot(data = data, x = "Transmission Type", y = "Popularity", ax = axes[1])
plt.sca(axes[1])
plt.ylabel('Popularity', fontsize = 18)
plt.xlabel('Transmission Type', fontsize = 18)
plt.xticks(rotation = 15);
```

```python
# Bar plot to observe the Wheel drive with MSRP and Popularity

fig, axes = plt.subplots(1, 2, figsize = (15, 6))
sns.set_style("white")

sns.barplot(data = data, x = "Driven_Wheels", y = "MSRP", ax = axes[0])
plt.sca(axes[0])
plt.ylabel('MSRP', fontsize = 18)
plt.xlabel('Wheel drive', fontsize = 18)
plt.xticks(fontsize = 15);

sns.barplot(data = data, x = "Driven_Wheels", y = "Popularity", ax = axes[1])
plt.sca(axes[1])
plt.ylabel('Popularity', fontsize = 18)
plt.xlabel('Wheel drive', fontsize = 18)
plt.xticks(fontsize = 15);
```

```python
# Bar plot to observe the Vehicle Style with MSRP

plt.figure(figsize = (12,10))
sns.set_style("white")
sns.barplot(data = data, y = "Vehicle Style", x = "MSRP")
plt.ylabel('Vehicle Style', fontsize = 20)
plt.xlabel('MSRP', fontsize = 20)
plt.show()
```

```python
# Line plot for Highway MPG and MSRP

plt.figure(figsize = (10,8))
sns.lineplot(data = data, x = "highway MPG", y = "MSRP", hue = "Transmission Type");
```

```python
# Let's see the most significant variable that affects the price

plt.figure(figsize = (8,8))
sns.set_style("darkgrid")
sns.scatterplot(data = data, y = "Engine HP", x = "MSRP");
```

```python
# Correlation Plot

fig, ax = plt.subplots(figsize = (10, 8))

sns.heatmap(data.corr(), annot = True, annot_kws = {'fontsize': 16}, ax = ax);
```
