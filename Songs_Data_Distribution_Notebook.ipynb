## Importing required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import StandardScaler
```

## Loading the required Dataset

```python
data = pd.read_csv('song_data.csv')
data.head()
```

```python
data.columns
```

## Fetching Continuous varibales 

```python
continuous = data.drop(['song_name','key','audio_mode','time_signature'], axis=1)
continuous
```

## Scaling the Continuous Variables 

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(continuous)
scaled
```

```python
continuous[list(continuous.columns)] = scaled
continuous.head()
```

## Creating a User defined function to check Normality

```python
def normality_check(var):
  a = stats.ks_1samp(x = var, cdf = stats.norm.cdf)
  print('According to KS test, the p-value for the given variable is {}'.format(a[1]))
  if a[1] < 0.5:
    print('Since the p-value is < 0.5, the given variable follows Normal distribution.')
  else:
    print('Since the p-value is < 0.5, the given variable does not follows Normal distribution.')
  print('---------------------------------------------------------------------')
  sns.set_style('whitegrid')
  plt.figure(figsize=(8,8))
  ax = sns.histplot(var, kde='True', bins=20)
  ax.lines[0].set_color('crimson')
  b = var.skew()
  print('The skewness value of the given variable is {}'.format(b))
  if b == 0:
    print('The distribution of the variable is perfect Normal.')
  elif 0<b<=1:
    print('''The skewness of the distribution is slightly positive.
(i.e) There is slightly more weight on the left side of the distribution.''')
  elif 1<b:
    print('''The skewness of the distribution is highly positive.
(i.e) There is more weight on the left side of the distribution.''')
  elif 0>b>= -1:
    print('''The skewness of the distribution is slightly negative.
(i.e) There is slightly more weight on the right side of the distribution.''')
  elif -1>b:
    print('''The skewness of the distribution is highly negative.
(i.e) There is more weight on the right side of the distribution.''')
```
